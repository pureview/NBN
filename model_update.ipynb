```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Perceptron
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
```

```python
url = pd.read_csv('data2.csv')
```

```python
url_df = pd.DataFrame(url)
url_df = url_df.sample(frac = 1)
display(url_df)
```

```python
url_arr = np.array(url_df)
print(url_arr)
```

```python
urls = [i[0] for i in url_arr]
```

```python
types = [i[1] for i in url_arr]
```

```python
#create 3-gram function
def getNGrams(query):
 
    tempQuery = str(query)
 
    ngrams = []
 
    for i in range(0,len(tempQuery)-3):
 
        ngrams.append(tempQuery[i:i+3])
 
    return ngrams
```

```python
tfidf_model = TfidfVectorizer(tokenizer=getNGrams)
sparse_result = tfidf_model.fit_transform(urls)
```

```python
print(sparse_result)
```

```python
X_train, X_test, y_train, y_test = train_test_split(sparse_result, types, test_size=0.2, random_state=42)
```

```python
'''

N_estimators: the maximum number of iterations of the weak learner
Oob_score : Whether to use the sample outside the bag to evaluate the quality of the model, using the data outside the bag to verify the decision tree model is a simple cross-validation. Low performance, but good results
Criterion: A criterion for evaluating features when dividing. One criterion is the Gini coefficient, and another optional criterion is information gain.

'''
#random_forest = RandomForestClassifier(n_estimators=100) 
random_forest = RandomForestClassifier(n_estimators=50,oob_score=True, criterion='gini', 
min_samples_split=2, min_samples_leaf=1)

random_forest.fit(X_train, y_train)
score = random_forest.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```

```python
#mlp = MLPClassifier()
mlp = MLPClassifier(hidden_layer_sizes=(50, ), activation='relu',solver='adam',
alpha=0.0005,max_iter=150,tol=0.0005, early_stopping=True, validation_fraction=0.1)

mlp.fit(X_train, y_train)
score = mlp.score(X_test, y_test)
print("score: {0:.2f} %".format(100 * score))
```
